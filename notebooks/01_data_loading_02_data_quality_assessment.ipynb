#### Handling Missing Values

```python
# Using forward fills within each ticker group to propagate existing values along the time dimension

df_i = df_i.groupby('ticker', group_keys=False).apply(lambda x: x.ffill())
```

```python
df_i.isnull().mean()*100
```

```python
# Using backward fill to replace remaining leading NaNs at the start of each stockâ€™s record

df_i = df_i.groupby('ticker', group_keys=False).apply(lambda x: x.bfill())
```

```python
df_i.isnull().sum()
```

```python
# To check for negative value in the numerical columns

nc_i = df_i.select_dtypes(include='number').columns
{c: (df_i[c] < 0).sum() for c in nc_i}
```

```python
# Sanity checks for technical indicators in df_i

checks = {
    'RSI range valid': ((df_i['rsi_14'] >= 0) & (df_i['rsi_14'] <= 100)).all(),
    'ATR positive': (df_i['atr_14'] >= 0).all(),
    'Volatility positive': (df_i['volatility_20'] >= 0).all(),
    'BB logic': (df_i['bb_upper'] > df_i['bb_middle']).all() and (df_i['bb_middle'] > df_i['bb_lower']).all(),
    'Price/SMA ratio reasonable': df_i['price_to_sma_50'].between(0.5, 1.5).all()
}

for k, v in checks.items():
    print(f"{k}: {'âœ… Passed' if v else 'âš ï¸ Check needed'}")
```

```python
df_i[df_i['price_to_sma_50'] > 1.5]
```

```python
# To check for outliers

iqr_outliers = {}

for col in nc_i:
    Q1 = df_i[col].quantile(0.25)
    Q3 = df_i[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    count = ((df_i[col] < lower) | (df_i[col] > upper)).sum()
    if count > 0:
        iqr_outliers[col] = count

iqr_outliers
```

### Observation of Findings

The Stock Prices with Indicators dataset (df_i) has been fully cleaned and validated. There were no duplicates but some missing values which were structural and valid from the computations. This was resolved using forward and backward filling within each ticker group to preserve time continuity and avoid unneccessary data loss while ensuring that no NaN values remained for modeling.

After forward-filling missing values within each ticker group, most NaNs were resolved, the few remaining NaNs were located at the beginning of each series, as expected due to indicator initialization.
These were successfully filled using a subsequent backward fill (bfill). This two-step fill strategy ensured complete data coverage while maintaining temporal consistency.

```python
# per-ticker date continuity

# Group by ticker and find any missing or extra trading days
def check_gaps(x):
    x = pd.DatetimeIndex(x)  # convert the group to DatetimeIndex
    expected_idays = valid_days[(valid_days >= x.min()) & (valid_days <= x.max())]
    missing_idays = expected_idays.difference(x)  # valid trading days missing in the data
    extra_idays = x.difference(expected_idays)    # days that exist in the data but not valid
    return pd.Series({'missing_days': missing_idays, 'extra_days': extra_idays})
```

```python
# Show tickers with missing or extra trading days

# Show only offenders (very simple printout)
for t in gaps.index.levels[0]:
    md = gaps.loc[(t, 'missing_days')]
    ed = gaps.loc[(t, 'extra_days')]
    if len(md) or len(ed):
        print(t)
        if len(md): print("  Missing:", list(md))
        if len(ed): print("  Extra:  ", list(ed))
```

```python
gaps = df_i.groupby('ticker')['date'].apply(check_gaps)

gaps
```

```python
# Count how many invalid trading days (extra) each ticker has
extra_count = df_i.groupby('ticker')['date'].apply(
    lambda x: len(pd.DatetimeIndex(x).difference(valid_days))
)

# Show only tickers with any extras
extra_count = extra_count[extra_count > 0]
extra_count
```
